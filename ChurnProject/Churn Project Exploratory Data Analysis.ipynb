```python
dataset_copy = dataset.copy()
```

```python
dataset_copy.TotalCharges = pd.to_numeric(dataset_copy.TotalCharges, errors='coerce')
```

```python
dataset_copy.describe()
```

```python
dataset_copy.isnull().sum()
```

```python
dataset_copy.loc[dataset_copy ['TotalCharges'].isnull() == True]
```

```python
dataset_copy.dropna(how = 'any', inplace = True)
dataset_copy.isnull().sum()
```

```python
print(dataset_copy['tenure'].max()) 
```

```python
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

dataset_copy['tenure_group'] = pd.cut(dataset_copy.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
dataset_copy['tenure_group'].value_counts()
```

```python
dataset_copy.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
```

```python
dataset_copy.head()
```

```python
for i, predictor in enumerate(dataset_copy.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=dataset_copy, x=predictor, hue='Churn')
```

```python
dataset_copy['Churn'] = np.where(dataset_copy.Churn == 'Yes',1,0)
```

```python
dataset_copy.head()
```

```python
dataset_copy_dummies = pd.get_dummies(dataset_copy)
dataset_copy_dummies.head()
```

```python
sns.lmplot(data=dataset_copy_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

```python
Mth = sns.kdeplot(dataset_copy_dummies.MonthlyCharges[(dataset_copy_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(dataset_copy_dummies.MonthlyCharges[(dataset_copy_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

```python
Tot = sns.kdeplot(dataset_copy_dummies.TotalCharges[(dataset_copy_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(dataset_copy_dummies.TotalCharges[(dataset_copy_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```
